```python
import pandas as pd
import os
import matplotlib.pyplot as plt

file_path=("C:/Users/krish/OneDrive/Desktop/Class_Activities/Project/Crime_Data.csv")
crime_data= pd.read_csv(file_path)
crime_data

```

```python
#Dropping the columns 
filterd_column_data=crime_data.drop(['DR_NO','AREA','Crm Cd 1',
       'Crm Cd 2', 'Crm Cd 3','Cross Street','Crm Cd','Weapon Used Cd','Rpt Dist No','Part 1-2', 'Crm Cd 4', 'Mocodes', 'Premis Cd', 'Status','Status Desc'], axis=1)
filterd_column_data
```

```python
#Data after dropping the Nan in "Weapon Desc"
filterd_column_data = filterd_column_data.dropna(subset=['Weapon Desc'])
filterd_column_data
```

```python
#Renaming all the columns 
clean_crime_data = filterd_column_data.rename(columns={
    "DATE OCC": "Date of Crime",
    "DATE RPTD": "Date Reported",
    "TIME OCC": "Time of Crime",
    "AREA NAME": "Neighborhood",
    "Crm Cd Desc": "Type of Crime",
    "Vict Age": "Age",
    "Vict Sex": "Gender",
    "Vict Descent": "Ethnicity",
    "Premis Desc": "Location",
    "Weapon Desc": "Weapon Used",
    "LOCATION": "Address",
    "LAT": "Latitude",
    "LON": "Longitude"
})
clean_crime_data
```

```python
#Again renaming the Ethnicity column in to meaningful 'H'- Hispanic, 'B'-Black,'W'-White,'O'-Others,'A'-Other Asians
clean_crime_data['Ethnicity'] = clean_crime_data['Ethnicity'].replace({
    'H': 'Hispanic',
    'B': 'Black',
    'W':'White',
    'O':'Others',
    'A':'Other Asians'})
clean_crime_data
```

```python
#Droppping the column Date Reported from dataframe

clean_crime_data.drop('Date Rptd', axis=1, inplace=True)
```

```python
clean_crime_data
```

```python
#Adding the month column to the dataframe
clean_crime_data['Month'] = pd.to_datetime(clean_crime_data['Date of Crime']).dt.month
clean_crime_data
```

```python
clean_crime_data.columns
```

```python
#Adding the column Hour to the dataframe
clean_crime_data['Hour'] = [time.hour for time in clean_crime_data['Time of Crime']]
clean_crime_data
```

```python
#Which neighbourhood area is having the highest and lowest crime
highest_crime_neighborhood = clean_crime_data['Neighborhood'].value_counts().idxmax()
highest_crime_tot=clean_crime_data['Neighborhood'].value_counts().max()

lowest_crime_neighborhood = clean_crime_data['Neighborhood'].value_counts().idxmin()
lowest_crime_tot=clean_crime_data['Neighborhood'].value_counts().min()

print(f"The neighborhood with the highest crime is: {highest_crime_neighborhood} with total of:{highest_crime_tot}")
print(f"The neighborhood with the lowest crime is: {lowest_crime_neighborhood} with total of:{lowest_crime_tot}")
```

```python
#Top 10 areas in the neighboorhood with highest crime
top5_high_crime_neighborhoods = clean_crime_data['Neighborhood'].value_counts().nlargest(10)
top5_high_crime_neighborhoods
```

```python
#Visualization 
import matplotlib.pyplot as plt

# Assuming top5_high_crime_neighborhoods is already calculated
top5_high_crime_neighborhoods = clean_crime_data['Neighborhood'].value_counts()

# Plotting the bar graph
plt.figure(figsize=(10, 4))
top5_high_crime_neighborhoods.plot(kind='bar', color='blue')
plt.title('The Neighborhood based Crime Counts')
plt.xlabel('Neighborhood')
plt.ylabel('Crime Counts')
plt.show()
```

```python
#Where do the Crime Occur- location based crime count
location_counts = clean_crime_data['Location'].value_counts()
location_counts
```

```python
import matplotlib.pyplot as plt

# Assuming top7_location_counts is already calculated
top7_location_counts = clean_crime_data['Location'].value_counts().nlargest(7)

plt.figure(figsize=(10, 10))
top7_location_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 7 Locations in Los Angeles with Highest Crime Counts')
plt.show()
```

```python
#Visualization of top places with highest crime
import folium

# Assuming top_location_counts is your Series
top_location_counts = clean_crime_data['Location'].value_counts().nlargest(10)

# Create a base map
crime_map = folium.Map(location=[34.0522, -118.2437], zoom_start=12, width=1000, height=800)

# Add markers for the top locations
for location, count in top_location_counts.items():
    location_data = clean_crime_data[clean_crime_data['Location'] == location]
    mean_lat = location_data['Latitude'].mean()
    mean_lon = location_data['Longitude'].mean()
    folium.Marker(
        location=[mean_lat, mean_lon],
        popup=f"{location}: {count} crimes",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(crime_map)

# Display the map
crime_map
```
